```python
from random import randint
```

```python
class Die:
    def __init__(self,numside=6):
        self.numside=numside
    def roll(self):
        return randint(1,self.numside)
```

```python
die=Die()
results=[]
for i in range(1000):
    result=die.roll()
    results.append(result)
#print(results)   
```

```python
freq=[]
for val in range(1,die.numside+1):
    f=results.count(val)
    freq.append(f)
    
print(freq)
```
